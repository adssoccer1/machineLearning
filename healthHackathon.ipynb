```
import pandas as pd

#Getting Data
url = 'https://raw.githubusercontent.com/adssoccer1/machineLearning2019/master/post-operative.csv'
names = ['CoreTemp', 'SurfTemp', 'OxygenSaturation', 'BloodPressure', 'SurfStability', 'CoreStability', 'BloodPressureStability', 'Comfort', 'Decision']
dataSet = pd.read_csv(url, names=names)
print(dataSet.head(10))
#dataSet.describe()


#helper function
def getAverage(list): 
  counter = 0
  sum = 0
  for instance in list: 
    if(instance == -1111):
      continue
    else:
      sum += instance
      counter += 1
  return sum/counter

CoreTemp = dataSet["CoreTemp"].copy()
newCoreTempList = []
for instance in CoreTemp: 
  #print(type(instance))
  if instance == ("low"): 
    newCoreTempList.append(1)
  elif instance == "mid":
    newCoreTempList.append(2)
  elif instance == "high":
    newCoreTempList.append(3)
  else: 
    print("error in newCoreTempList ", instance)
print(newCoreTempList[0:9])
dataSet["CoreTemp"] = newCoreTempList

SurfTemp = dataSet["SurfTemp"].copy()
newSurfTempList = []
for instance in SurfTemp: 
  if instance == ("low"): 
    newSurfTempList.append(1)
  elif instance == "mid":
    newSurfTempList.append(2)
  elif instance == "high":
    newSurfTempList.append(3)
  else: 
    print("error in newSurfTempList ", instance)
print(newSurfTempList[0:9])
dataSet["SurfTemp"] = newSurfTempList

OxSat = dataSet["OxygenSaturation"].copy()
newOxSatList = []
for instance in OxSat:
  if instance == "excellent": 
    newOxSatList.append(3)
  elif instance == "good":
    newOxSatList.append(2)
  else: 
    print("error in newStatLsit", instance)
dataSet["OxygenSaturation"] = newOxSatList 

BloodPressure = dataSet["BloodPressure"].copy()
newBloodPressureList = []
for instance in BloodPressure:
  if instance == ("low"): 
    newBloodPressureList.append(1)
  elif instance == "mid":
    newBloodPressureList.append(2)
  elif instance == "high":
    newBloodPressureList.append(3)
  else: 
    print("error in newBloodPressureList ", instance)
print(len(newBloodPressureList))
dataSet["BloodPressure"] = newBloodPressureList

SurfStability = dataSet["SurfStability"].copy()
newSurfStabilitylist = []
for instance in SurfStability:
  if instance == ("stable"): 
    newSurfStabilitylist.append(2)
  elif instance == "unstable":
    newSurfStabilitylist.append(0)
  else: 
    print("error in newSurfStabilitylist ", instance)
dataSet["SurfStability"] = newSurfStabilitylist

CoreStability = dataSet["CoreStability"].copy()
newCoreStabilityList = []
for instance in CoreStability:
  if instance == ("stable"): 
    newCoreStabilityList.append(2)
  elif instance == "unstable":
    newCoreStabilityList.append(0)
  elif instance == "mod-stable":
    newCoreStabilityList.append(1)
  else: 
    print("error in newCoreStabilityList ", instance)
dataSet["CoreStability"] = newCoreStabilityList

BloodPressureStability = dataSet["BloodPressureStability"].copy()
newBloodPressureStabilityList = []
for instance in BloodPressureStability:
  if instance == ("stable"): 
    newBloodPressureStabilityList.append(2)
  elif instance == "unstable":
    newBloodPressureStabilityList.append(0)
  elif instance == "mod-stable":
    newBloodPressureStabilityList.append(1)
  else: 
    print("error in newBloodPressureStabilityList ", instance)
dataSet["BloodPressureStability"] = newBloodPressureStabilityList


Comfort = dataSet["Comfort"].copy()
newComfortList = []
for instance in Comfort:
  if instance == "?":
    newComfortList.append(-1111)
  else:
    newComfortList.append(int(instance))
average = getAverage(newComfortList)
for index in range(len(newComfortList)):
  if newComfortList[index] == -1111:
    newComfortList[index] = average
dataSet["Comfort"] = newComfortList 

Decision = dataSet["Decision"].copy()
newDecisionList = []
for instance in Decision:
  if type(instance) != str:
    instance = str(instance)
  if instance == "A":
    newDecisionList.append(1)
  elif instance == "S":
    newDecisionList.append(2)
  elif instance == "I":
    newDecisionList.append(0)
  else:
    print("error in newDecisionList ", instance)
dataSet["Decision"] = newDecisionList

print(dataSet.head(10))
```

```
corr_matrix = dataSet.corr()
print(corr_matrix)

#split dataSet and Decision to prepare for standard scaling and train_test_split
Decision = dataSet["Decision"].copy()
dataSetTwo = dataSet.drop("Decision", axis=1)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(dataSetTwo)
StandardScaler(copy=True, with_mean=True, with_std=True)
scaler.mean_
dataSetTwo = scaler.transform(dataSetTwo)


from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(dataSetTwo, Decision, test_size=0.1, random_state=42)

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
import numpy as np

lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
y_pred = lin_reg.predict(X_test)
lin_mse = mean_squared_error(y_test, y_pred)
print("LINEAR REGRESSION")
print("mean squared error: " , lin_mse)
lin_rmse = np.sqrt(lin_mse)
print("root mean squared error: " , lin_rmse)
lin_mae = mean_absolute_error(y_test, y_pred)
print("mean absolute error: " , lin_mae, "\n")

from sklearn.tree import DecisionTreeRegressor
tree_reg = DecisionTreeRegressor(random_state=42)
tree_reg.fit(X_train, y_train)
y_pred = tree_reg.predict(X_test)
tree_mse = mean_squared_error(y_test, y_pred)
print("DECISION TREE")
print("mean squared error: " , tree_mse)
tree_rmse = np.sqrt(tree_mse)
print("root mean squared error: " , tree_rmse)
tree_mae = mean_absolute_error(y_test, y_pred)
print("mean absolute error: " , tree_mae, "\n")
#cross validation on the decision tree model

from sklearn.ensemble import RandomForestRegressor
forest_reg = RandomForestRegressor(n_estimators=100, random_state=42)
forest_reg.fit(X_train, y_train)
y_pred = forest_reg.predict(X_test)
forest_mse = mean_squared_error(y_test, y_pred)
print("RANDOM FOREST")
print("mean squared error: " , forest_mse)
forest_rmse = np.sqrt(forest_mse)
print("root mean squared error: " , forest_rmse)
forest_mae = mean_absolute_error(y_test, y_pred)
print("mean absolute error: " , forest_mae, "\n")

#Create interface to get input from a user
print()
```
